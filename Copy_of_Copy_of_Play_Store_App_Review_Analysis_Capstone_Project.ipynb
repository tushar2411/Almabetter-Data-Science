#*In Rating column we have some null values. To remove these values we'll calculate median of this column by creating function*

```python
def cal_median(counts):
  return counts.fillna(counts.median)
```

```python
playdataF.Rating = playdataF['Rating'].transform(cal_median)
```

#*Now we will apply this function for same the 'Content Rating,Current Ver,Android Ver, Type'*

```python
playdataF.Current_Ver = playdataF['Current_Ver'].transform(cal_median)
```

```python
playdataF.Android_Ver = playdataF['Android_Ver'].transform(cal_median)
```

```python
playdataF.Type = playdataF['Type'].transform(cal_median)
```

#*We can also see that above information about data there is a Price column which has all numeric float values plus there is $ sign we will remove it and convert into float values.'*

```python
playdataF['Price'].unique()
```

```python
for col in range(0,len(playdataF['Price'])):
  if '$' in playdataF.loc[col,'Price']:
      playdataF.loc[col,'Price'] = playdataF.loc[col,'Price'][1:]
  playdataF.loc[col,'Price'] =  float(playdataF.loc[col,'Price'])
```
